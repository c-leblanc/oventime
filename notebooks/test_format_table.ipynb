```python
raw_generation = pd.read_csv("../data/raw/generation.csv")
raw_generation = raw_generation.drop(raw_generation.index[0])
raw_generation = raw_generation.rename(columns=
                                       {"Unnamed: 0":"time",
                                        "Biomass":"Biomass",
                                        "Energy storage":"Storage_PROD",
                                        "Energy storage.1": "Storage_CONS",
                                        "Fossil Gas":"Gas",
                                        "Fossil Hard coal":"Coal",
                                        "Fossil Oil":"Oil",
                                        "Hydro Pumped Storage":"PHS_PROD",
                                        "Hydro Pumped Storage.1":"PHS_CONS",
                                        "Hydro Run-of-river and poundage":"RoR",
                                        "Hydro Water Reservoir":"HydroDams",
                                        "Nuclear":"Nuclear",
                                        "Solar":"Solar",
                                        "Waste":"Waste",
                                        "Wind Offshore":"Wind_off",
                                        "Wind Onshore":"Wind_on"})

gen_technos = raw_generation.columns.drop("time")
raw_generation[gen_technos] = raw_generation[gen_technos].apply(pd.to_numeric, errors="coerce")

raw_generation["Storage"] = raw_generation["Storage_PROD"] - raw_generation["Storage_CONS"]
raw_generation = raw_generation.drop(columns=["Storage_PROD", "Storage_CONS"])
raw_generation["PHS"] = raw_generation["PHS_PROD"] - raw_generation["PHS_CONS"]
raw_generation = raw_generation.drop(columns=["PHS_PROD", "PHS_CONS"])

raw_generation
```
